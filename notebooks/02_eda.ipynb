```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_data(file_path):
    return pd.read_json(file_path, lines=True)

file_path = "../data/MLA_100k_checked_v3.jsonlines"
df = load_data(file_path)
```

```python
# Calcular valores faltantes
missing_data = df.isnull().sum()
total_rows = len(df)

# Crear resumen con porcentaje y tipo de dato
missing_summary = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': missing_data.values,
    'Missing_Percentage': (missing_data.values / total_rows) * 100,
    'Data_Type': df.dtypes.values
})

# Identificar tipos de columnas
numerical_columns = df.select_dtypes(include=[np.number]).columns
categorical_columns = df.select_dtypes(include=['object']).columns
datetime_columns = df.select_dtypes(include=['datetime64[ns]', 'datetime64[ns, UTC]']).columns

# Clasificar tipo de variable
missing_summary['Variable_Type'] = missing_summary['Column'].apply(
    lambda x: (
        'Numerical' if x in numerical_columns else
        'Categorical' if x in categorical_columns else
        'Datetime' if x in datetime_columns else
        'Other'
    )
)

# Ordenar por mayor cantidad de valores faltantes
missing_summary = missing_summary.sort_values(by='Missing_Count', ascending=False).reset_index(drop=True)

# Mostrar dimensiones del dataset
print("Dimensiones del dataset:")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
print('--------')

# Conteo absoluto y porcentaje por categoría en 'condition'
condition_counts = df["condition"].value_counts()
condition_percentages = df["condition"].value_counts(normalize=True) * 100

# Unir ambos en un solo DataFrame
condition_summary = pd.DataFrame({
    "Count": condition_counts,
    "Percentage": condition_percentages.round(2)
})
print("Valores distintos en variable 'condition':")
print(condition_summary)
print('--------')

# Mostrar resumen de valores faltantes y tipos
print("Resumen de valores faltantes y tipos de variables:")
# display(missing_summary)
missing_summary
```

```python
# Visualizar la distribución de la variable objetivo 'condition'
plt.figure(figsize=(6,4))
sns.countplot(x="condition", data=df, palette='inferno')
plt.title("Distribución de la variable objetivo: condition")
plt.show()
```

La variable objetivo `condition` presenta una distribución balanceada entre ítems `nuevos` y `usados`, lo cual es ideal para modelado supervisado. Esta proporción cercana al 50/50 permite utilizar **accuracy** como métrica principal sin riesgo de sesgo por desbalance de clases, asegurando una evaluación justa del rendimiento del modelo.

```python
# Visualizar la distribución de la variable objetivo 'condition' con log scale
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="condition", y="price", palette='inferno')
plt.yscale("log")
plt.title("Precio según condición (log)")
plt.show()
```

La gráfica muestra que, en términos generales los productos `nuevos` tienen precios medianos más altos que los `usados`. Sin embargo, ambos presentan una alta dispersión y valores atípicos extremos. El uso de escala logarítmica revela mejor la concentración de precios en rangos bajos, destacando que aunque hay productos muy costosos en ambas categorías, son excepcionales. Esto sugiere que la `condición` del producto influye en el `precio`, pero no es el único factor determinante.

```python
# Visualizar la distribución de precios por condición
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='price', hue='condition', log_scale=True, bins=50, palette='inferno', kde=True)
plt.title('Distribución de precios por condición')
plt.show()
```

El gráfico muestra que los productos `nuevos` tienden a tener precios más altos y mayor dispersión que los `usados`, con un claro sesgo a la derecha en ambas distribuciones. Usar una escala logarítmica permite evidenciar el solapamiento en rangos bajos, pero también cómo los productos `nuevos` dominan en precios más altos.

```python
# Location 
df['state_id'] = df['seller_address'].apply(lambda x: x['state']['id'] if isinstance(x, dict) else 'unknown')

# Visualizar la distribución de precios por provincia
top_states = df['state_id'].value_counts().nlargest(10).index
plt.figure(figsize=(10, 6))
sns.countplot(data=df[df['state_id'].isin(top_states)], y="state_id", hue="condition", order=top_states, palette='inferno')
plt.title("Top 10 provincias por cantidad de publicaciones")
plt.show()
```

La mayoría de las publicaciones provienen de las provincias `AR-C` y `AR-B`, concentrando ampliamente la oferta tanto de productos nuevos como usados. `AR-C` lidera en ambos casos, con ligera ventaja de productos nuevos. A partir de la tercera provincia, la cantidad de publicaciones disminuye drásticamente. Esto sugiere una fuerte centralización del mercado en regiones específicas, lo cual podría estar relacionado con densidad poblacional o infraestructura comercial.

```python
# Shipping info
df['is_free_shipping'] = df['shipping'].apply(lambda x: x.get('free_shipping', False) if isinstance(x, dict) else False).astype(int)
df['shipping_mode'] = df['shipping'].apply(lambda x: x.get('mode', 'unknown') if isinstance(x, dict) else 'unknown')
df['local_pick_up'] = df['shipping'].apply(lambda x: x.get('local_pick_up', False) if isinstance(x, dict) else False).astype(int)

# Gráficos de shipping
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Free shipping
sns.countplot(ax=axes[0], x="is_free_shipping", hue="condition", data=df, palette='inferno')
axes[0].set_title("Free shipping vs Condition")
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(["No", "Sí"])

# Shipping mode
sns.countplot(ax=axes[1], x="shipping_mode", hue="condition", data=df, 
              order=df["shipping_mode"].value_counts().index, palette='inferno')
axes[1].set_title("Modo de envío por condición del producto")
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)

# Local pick-up
sns.countplot(ax=axes[2], x="local_pick_up", hue="condition", data=df, 
              palette='inferno', hue_order=["new", "used"])
axes[2].set_title("Retiro en persona vs Condition")
axes[2].set_xticks([0, 1])
axes[2].set_xticklabels(["No", "Sí"])

plt.tight_layout()
plt.show()
```

Las variables derivadas de `shipping` muestran diferencias relevantes según la condición del producto. El envío gratis (poco comun) y el modo `me2` están más asociados a productos nuevos, mientras que los usados predominan en métodos no especificados. Además, el retiro en persona es más frecuente en nuevos, lo cual sugiere patrones logísticos diferentes según el estado del ítem.

```python
# Agrupar datos por día y condición para suavizar la curva
df['start_date'] = df['start_time'].dt.date
daily_price = df.groupby(['start_date', 'condition'])['price'].mean().reset_index()

# Plot
plt.figure(figsize=(14, 6))
sns.lineplot(data=daily_price, x='start_date', y='price', hue='condition', palette='inferno')
plt.title("Tendencia de precio medio por día y condición")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Este gráfico muestra la **tendencia del precio medio diario** por condición del producto `nuevo` o `usado`. Se observan múltiples **picos atípicos**, especialmente entre los productos `usados`, lo que sugiere la presencia de **outliers** que inflan artificialmente el promedio en ciertos días. A pesar del ruido, se percibe que ambos tipos de productos tienden a mantener precios promedio bajos.

```python
# Extraemos año y mes
df['year_month'] = df['date_created'].dt.to_period('M').astype(str)

# Agrupamos y contamos publicaciones por año-mes
monthly_counts = df.groupby('year_month').size().reset_index(name='publications')

# Orden cronológico (por si acaso)
monthly_counts = monthly_counts.sort_values('year_month')

# Gráfico
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_counts, x='year_month', y='publications', marker='o', color='purple')
plt.title('Cantidad de publicaciones por mes y año')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Este gráfico muestra la evolución mensual de publicaciones en el marketplace. Se observa una **actividad casi nula** entre 2013 y mediados de 2015. A partir de **julio de 2015**, hay un **aumento drástico** y sostenido en el volumen diario. Esto sugiere una **carga masiva de datos o una expansión del sistema** en ese periodo, marcando un antes y un después en el comportamiento del marketplace.

```python
# Payment methods
df['payment_method_group'] = df['non_mercado_pago_payment_methods'].apply(lambda x: len(x) if isinstance(x, list) else 0).apply(lambda n: n if n <= 3 else '4+')

# Gráfico de métodos de pago
plt.figure(figsize=(7,4))
sns.countplot(x="payment_method_group", hue="condition", data=df, order=["0", "1", "2", "3", "4+"], palette='inferno')
plt.title("Cantidad de métodos de pago alternativos por condición")
plt.show()
```

Los productos `nuevos` suelen ofrecer una mayor diversidad de `métodos de pago` no integrados con MercadoPago, especialmente en los grupos de 2 y 3 opciones. Esta diversidad puede reflejar una estrategia comercial más estructurada en terminos de accesibilidad y flexibilidad del vendedor.

```python
# Tags
df['has_dragged_visits'] = df['tags'].apply(lambda x: 'dragged_visits' in x or 'dragged_bids_and_visits' in x if isinstance(x, list) else False).astype(int)
df['has_good_thumbnail'] = df['tags'].apply(lambda x: 'good_quality_thumbnail' in x if isinstance(x, list) else False).astype(int)
df['has_poor_thumbnail'] = df['tags'].apply(lambda x: 'poor_quality_thumbnail' in x if isinstance(x, list) else False).astype(int)
df['was_relisted'] = df['tags'].apply(lambda x: 'free_relist' in x if isinstance(x, list) else False).astype(int)

# Gráficos de tags
tag_cols = ['has_dragged_visits', 'has_good_thumbnail', 'has_poor_thumbnail', 'was_relisted']
titles = ['Arrastró visitas', 'Buena miniatura', 'Miniatura deficiente', 'Fue republicado']

fig, axes = plt.subplots(2, 2, figsize=(14, 8))

for ax, col, title in zip(axes.flatten(), tag_cols, titles):
    sns.countplot(ax=ax, x=col, hue="condition", data=df, palette='inferno', hue_order=["new", "used"])
    ax.set_title(f"{title} vs condición")
    ax.set_xticks([0, 1])
    ax.set_xticklabels(["No", "Sí"])

plt.tight_layout()
plt.show()
```

Las etiquetas del anuncio revelan patrones útiles. La presencia de visitas arrastradas es más frecuente en productos nuevos, lo que sugiere mayor exposición. Las miniaturas de buena calidad, aunque poco comunes, aparecen más en artículos nuevos, mientras que las de baja calidad son casi inexistentes y poco discriminativas. Por último, los productos nuevos tienden a no ser republicados con mayor frecuencia, lo que puede reflejar mayor demanda. Estas variables capturan aspectos indirectos del comportamiento del vendedor y del mercado.

```python
# Visualizar la relación entre variables numéricas
cols_to_plot = ['price', 'sold_quantity', 'initial_quantity', 'available_quantity']
# Filtrar columnas numéricas
df_pair = df[cols_to_plot + ['condition']].copy()
# Convertir a logaritmo (evitar valores negativos)
df_pair[cols_to_plot] = df_pair[cols_to_plot].apply(lambda x: np.log1p(x))

# Crear un gráfico de dispersión
sns.pairplot(df_pair, hue='condition', palette='inferno')
plt.suptitle("Relaciones entre variables numéricas (escala log)", y=1.02)
plt.show()
```

Este pairplot muestra las relaciones entre variables numéricas clave en escala logarítmica: `price`, `sold_quantity`, `initial_quantity` y `available_quantity`, diferenciadas por condición del producto.

Se observan fuertes acumulaciones en valores bajos (especialmente en `sold_quantity`), lo que indica muchos productos sin ventas. La diagonal entre `initial_quantity` y `available_quantity` sugiere que muchos productos no se han vendido. Las correlaciones son más visibles entre cantidades (`initial_quantity` vs `available_quantity`) que con el precio. La diferencia por condición es mínima, indicando comportamientos similares entre productos nuevos y usados en estas variables.

```python
# Definir condiciones
conditions = df['condition'].unique()
num_conditions = len(conditions)

# Crear subplots
fig, axes = plt.subplots(1, num_conditions, figsize=(6 * num_conditions, 5))

# Si solo hay una condición, convertir axes en lista para compatibilidad
if num_conditions == 1:
    axes = [axes]

# Graficar cada heatmap
for i, cond in enumerate(conditions):
    subset = df[df['condition'] == cond][['price', 'sold_quantity', 'initial_quantity', 'available_quantity']]
    corr = subset.apply(np.log1p).corr()

    sns.heatmap(corr, annot=True, cmap='viridis', fmt=".2f", ax=axes[i])
    axes[i].set_title(f"Correlación (log) - {cond.capitalize()}")

plt.suptitle("Matrices de correlación por condición del producto", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

Las matrices muestran que el precio no guarda correlación significativa con otras variables, tanto en productos nuevos como usados. En cambio, existe una correlación casi perfecta entre `initial_quantity` y `available_quantity`, lo que sugiere baja rotación de inventario. `Sold_quantity` presenta correlaciones bajas con las demás variables, indicando que las ventas no dependen directamente del stock publicado. Estos patrones son consistentes entre condiciones, lo que sugiere un comportamiento estructural del marketplace más que diferencias por tipo de producto.
